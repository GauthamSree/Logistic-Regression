## Spliting the Data & Normalizing

```python
train_idx = round(len(df)*0.6)
valid_idx = train_idx + (len(df) - train_idx)//2
```

```python
idxs  = np.random.permutation(len(df))
```

```python
train, valid, test = df.iloc[idxs[:train_idx]], df.iloc[idxs[train_idx:valid_idx]], df.iloc[idxs[valid_idx:]]
```

```python
train.shape, valid.shape, test.shape
```

```python
train_mean = train.drop('diagnosis', axis = 1).mean()
train_std = train.drop('diagnosis', axis = 1).std()
```

```python
def normalize(df_1, mean, std, skip_fld=None):
    df = df_1.copy()
    if skip_fld is None:
        df = (df - mean)/(1e-7 + std)
    else:
        y = df[skip_fld]
        df.drop('diagnosis', axis = 1, inplace = True)
        df = (df - mean)/(1e-7 + std)
        df = pd.concat([df, y], axis = 1)
    return df
```

```python
train = normalize(train, train_mean, train_std, 'diagnosis')
valid = normalize(valid, train_mean, train_std, 'diagnosis')
test = normalize(test, train_mean, train_std, 'diagnosis')
```

```python
train.to_csv('data/train.csv', index=False)
valid.to_csv('data/valid.csv', index=False)
test.to_csv('data/test.csv', index=False)
```
